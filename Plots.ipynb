```python
import pandas as pd

file_path = "CleanedDataFrame.csv"

df = pd.read_csv(file_path)
```

```python
df['Average Salary'].hist(bins=10, color="Blue", edgecolor="Black")
```

```python
df[df['Average Salary'] > 200]['Average Salary'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.displot(x="Rating", data=df, bins=10)
```

```python
sns.displot(x="Average Salary", data=df, bins=10)
```

```python
sns.countplot(x="Job Title", data=df)
plt.xticks(rotation=90)
```

```python
sns.boxenplot(x="Rating", data=df)
```

```python
df = df.sort_values(by='Average Salary', ascending=True)
```

```python
df
```

```python
plt.barh(df['Job Title'], df['Average Salary'], color="skyBlue")
plt.xlabel('Average Salary')
plt.ylabel('Job Title')
plt.title('Average Salary by Job Title')
plt.grid(axis='x')
plt.show()
```

```python
plt.bar(df['State'], df['Average Salary'])
plt.xlabel('State')
plt.ylabel('Average Salary')
plt.title('Average Salary by State')
plt.grid(axis='x')
plt.xticks(rotation=45)
plt.show()
```
